Checking Missing Values

```python
df.isnull().sum()
```

```python
df['gender'].fillna('Unknown', inplace=True)
#replacing missing values with "Unknown" in "gender" column
```

```python
df['tenure'].fillna(df['tenure'].median(), inplace=True)
#replacing the missing values in "tenure" column with median
```

```python
df.isnull().sum()
#checking for missing values again
```

Analysis

```python
#Distribution of Gender
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Percentage distribution
gender_percent = df['gender'].value_counts(normalize=True) * 100
print(gender_percent)
```

```python
#Distribution of Tenure
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Tenure')
plt.xlabel('Tenure (Days since joining Facebook)')
plt.ylabel('Count')
plt.show()

print("Average tenure:", df['tenure'].mean())
```

```python
#Friend Count by Gender
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='friend_count', data=df, palette='coolwarm')
plt.title('Friend Count Distribution by Gender')
plt.show()

df.groupby('gender')['friend_count'].mean().sort_values(ascending=False)
```

```python
#Like vs Age
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='likes', data=df, alpha=0.5)
plt.title('Likes vs Age')
plt.xlabel('Age')
plt.ylabel('Number of Likes')
plt.show()
```

```python
#Correlation Between Numerical Columns
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
