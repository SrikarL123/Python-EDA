Univariate Analysis (Analyzing Single Variables)

```python
#Numerical Variables (Histograms)
# Set the style for plots
sns.set_style('whitegrid')

# Get all numerical column names
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

print("\n--- Plotting Histograms for Numerical Variables ---")
# Plot a histogram for each numerical variable
for col in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}', fontsize=15)
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.savefig(f'{col}_histogram.png')
    plt.show() # In Colab, this will display the plot
    plt.close() # Close the plot to avoid overlap in loops
```

Bivariate Analysis (Finding Relationships)

```python
#Correlation Heatmap (Numerical vs. Numerical)
print("\n--- Plotting Correlation Heatmap ---")
plt.figure(figsize=(12, 8))
# Calculate the correlation matrix
corr_matrix = df[numerical_columns].corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features', fontsize=15)
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()
plt.close()
```

Key Scatter Plots (Numerical vs. Exam Score)

```python
print("\n--- Plotting Key Relationships with Exam Score (Scatter Plots) ---")
# Define columns you want to plot against the exam score
key_cols_to_plot = ['study_hours_per_day', 'social_media_hours', 'attendance_percentage', 'sleep_hours']

for col in key_cols_to_plot:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df[col], y=df['exam_score'])
    # Add a regression line to see the trend
    sns.regplot(x=df[col], y=df['exam_score'], scatter=False, color='red')
    plt.title(f'Exam Score vs. {col}', fontsize=15)
    plt.xlabel(col)
    plt.ylabel('Exam Score')
    plt.savefig(f'exam_score_vs_{col}_scatterplot.png')
    plt.show()
    plt.close()
```
